```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
data_items = pd.read_csv('item.csv')
data_promotions = pd.read_csv('promotion.csv')
data_supermarket = pd.read_csv('supermarkets.csv')
data_sales = pd.read_csv('sales.csv')
```

```python
df_items = pd.DataFrame(data_items)
df_promotions = pd.DataFrame(data_promotions)
df_supermarkets = pd.DataFrame(data_supermarket)
df_sales = pd.DataFrame(data_sales)
```

```python
print(df_sales.duplicated())
df_sales.drop_duplicates(inplace = True)
```

### Exploratory Data Analysis

```python
# Merge sales and item datasets on 'code'
sales_item_merged = pd.merge(df_sales, df_items, how='left', on='code')

# Check the merged result
sales_item_merged.head()
```

```python
# Merge the sales-item merged dataset with promotion dataset on 'code' and 'week'
sales_item_promo_merged = pd.merge(sales_item_merged, df_promotions, how='left', on=['code', 'week'])

# Check the merged result
sales_item_promo_merged.head(25)
```

```python
# Merge with supermarket dataset on 'supermarket_No' from supermarkets.csv and 'supermarket' from sales_item_promo_merged
final_merged_df = pd.merge(sales_item_promo_merged, df_supermarkets, how='left', left_on='supermarket', right_on='supermarket_No')

# Check the merged result
print(final_merged_df.head())
```

```python
final_merged_df
```

```python
empty_values = final_merged_df.isnull().sum()
```

```python
print(empty_values)
```

```python
df_final = final_merged_df.drop(['supermarkets', 'feature', 'display', 'province_y','size'], axis=1)
```

### EDA

```python
# Group by supermarket and item type to analyze sales performance
sales_performance = df_final.groupby(['supermarket', 'type']).agg({
    'amount': 'sum',  # total sales
    'units': 'sum',   # total units sold
}).reset_index()

# View the grouped data
print(sales_performance.head(20))
```

```python
# Group by province and supermarket to see regional performance
regional_sales_performance = df_final.groupby(['province_x', 'supermarket']).agg({
    'amount': 'sum',
    'units': 'sum',
}).reset_index()

# View the grouped data
print(regional_sales_performance.head(20))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.barplot(data=df_final, x='type', y='amount', estimator=sum, ci=None)
plt.title('Total Sales by Item Type')
plt.ylabel('Total Sales Amount')
plt.xlabel('Item Type')
plt.show()
```

```python
# Total sales by brand
plt.figure(figsize=(10, 5))
sns.barplot(data=df_final, x='brand', y='amount', estimator=sum, ci=None)
plt.title('Total Sales by Brand')
plt.ylabel('Total Sales Amount')
plt.xlabel('Brand')
plt.xticks(rotation=45)
plt.show()
```

```python
daily_sales = df_final.groupby('day')['amount'].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=daily_sales, x='day', y='amount', marker='o')
plt.title('Total Sales by Day')
plt.ylabel('Total Sales Amount')
plt.xlabel('Day of the Month')
plt.xticks(range(1, 8))  # Adjust according to your data range
plt.grid()
plt.show()
```

```python
# Promotional Effectiveness Analysis (Assuming a 'promotional' column exists)
# Example data creation for promotional column

plt.figure(figsize=(10, 5))
sns.boxplot(data=df_final, x='voucher', y='amount')
plt.title('Sales Distribution: Promotional vs. Non-Promotional')
plt.xticks([0, 1], ['Non-Promotional', 'Promotional'])
plt.ylabel('Sales Amount')
plt.xlabel('Promotion Status')
plt.show()
```
